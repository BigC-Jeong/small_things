```python
name = input("이름을 입력하세요:")
age = int(input("나이를 입력하세요:"))
if age >= 19:
    print(name, "님은 성년입니다.")
    print("입장하세요~")
else :
  print(name,"님은 미성년입니다.")
  print("출입할 수 없습니다.")
```
